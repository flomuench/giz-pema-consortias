```python
#Using GoogleV3 worked, API key needed (300$ free requests per month)
#Ecommerce DATA
from geopy.geocoders import GoogleV3
df = pd.read_csv('ecommerce.csv')

google_api_key = 'API_KEY_HERE'  # Replace with your actual API key
geolocator = GoogleV3(api_key=google_api_key)

# Create new columns for latitude and longitude
df['latitude'] = 0.0
df['longitude'] = 0.0

# Geocode addresses and update the DataFrame
for i, address in enumerate(df['rg_adresse']):
    try:
        location = geolocator.geocode(address)
        if location is not None:
            df.at[i, 'latitude'] = location.latitude
            df.at[i, 'longitude'] = location.longitude
    except Exception as e:
        print(f"Geocoding failed for address at index {i}: {address}")
        print(f"Error: {e}")

df.to_csv('/content/ecommerce_geolocation.csv', index=False)
```
